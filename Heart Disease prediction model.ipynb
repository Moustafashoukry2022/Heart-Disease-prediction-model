#AI models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data_frame=pd.DataFrame(main_data)
data_frame
```

```python
# Use One-Hot Encoding for non-ordinal categorical variables
encoding_data = pd.get_dummies(data_frame, drop_first=True)  # 'drop_first=True' avoids dummy variable trap
encoding_data
```

```python
y=encoding_data['Heart Disease Status_Yes']
x=encoding_data.drop('Heart Disease Status_Yes',axis=1)
```

```python
from sklearn.impute import SimpleImputer

# Initialize SimpleImputer with median strategy
imputer = SimpleImputer(strategy='median')

# Impute missing values for all columns with NaN
x_imputed = imputer.fit_transform(x)

# Convert the result back to DataFrame if needed
x = pd.DataFrame(x_imputed, columns=x.columns)
```

```python
# Check which columns have NaN values
nan_columns = x.isna().sum()

# Display columns with NaN values and the number of NaNs in each column
print(nan_columns[nan_columns > 0])
```

```python
x
```

```python
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
